```python
import geopandas as gpd
import pandas as pd
import matplotlib.pyplot as plt

# Load the shapefile
shapefile_path = 'Admin2.shp'  # Replace with the actual shapefile name
india_states = gpd.read_file(shapefile_path)

# Load the CSV file
csv_path = 'Rainfall_Expectation.csv'  # Replace with the actual CSV file name
rainfall_data = pd.read_csv(csv_path)

# Merge shapefile and CSV data
india_states = india_states.merge(rainfall_data, left_on='ST_NM', right_on='ST_NM', how='left')

# Add a column for classification
india_states['Category'] = india_states['Rainfall_Expectation'].apply(lambda x: 'Positive' if x > 0 else 'Negative')

# Plot the map
fig, ax = plt.subplots(1, 1, figsize=(15, 12))
india_states.boundary.plot(ax=ax, linewidth=0.8, color='gray')  # Plot boundaries
india_states.plot(
    column='Category',
    cmap='RdYlBu',
    legend=True,
    ax=ax,
    legend_kwds={'title': 'Rainfall Expectation', 'fontsize': 10, 'loc': 'lower left', 'frameon': False},
)

# Add state names
for idx, row in india_states.iterrows():
    if row['geometry'] is not None:
        centroid = row['geometry'].centroid
        ax.text(centroid.x, centroid.y, row['ST_NM'], fontsize=8, ha='center', color='black')

# Customize the plot
ax.set_title('State-level Rainfall Expectation', fontsize=18, fontweight='bold')
ax.axis('off')  # Turn off axis
ax.set_facecolor("#f7f7f7")  # Set background color

# Show the plot
plt.tight_layout()
plt.show()
```
