```python
import geopandas as gpd
import pandas as pd
import matplotlib.pyplot as plt

# Load the shapefile
shapefile_path = 'Admin2.shp'  # Replace with the actual shapefile name
india_states = gpd.read_file(shapefile_path)

# Load the CSV file
csv_path = 'Rainfall_Expectation.csv'  # Replace with the actual CSV file name
rainfall_data = pd.read_csv(csv_path)

# Merge shapefile and CSV data
india_states = india_states.merge(rainfall_data, left_on='ST_NM', right_on='ST_NM', how='left')

# Add a column for classification
india_states['Category'] = india_states['Rainfall_Expectation'].apply(lambda x: 'Positive' if x > 0 else 'Negative')

# Plot the map
fig, ax = plt.subplots(1, 1, figsize=(15, 12))
india_states.boundary.plot(ax=ax, linewidth=0.8, color='gray')  # Plot boundaries
india_states.plot(
    column='Category',
    cmap='RdYlBu',
    legend=True,
    ax=ax,
    legend_kwds={'title': 'Rainfall Expectation', 'fontsize': 10, 'loc': 'lower left', 'frameon': False},
)

# Add state names
for idx, row in india_states.iterrows():
    if row['geometry'] is not None:
        centroid = row['geometry'].centroid
        ax.text(centroid.x, centroid.y, row['ST_NM'], fontsize=8, ha='center', color='black')

# Customize the plot
ax.set_title('State-level Rainfall Expectation', fontsize=18, fontweight='bold')
ax.axis('off')  # Turn off axis
ax.set_facecolor("#f7f7f7")  # Set background color

# Show the plot
plt.tight_layout()
plt.show()
```

```python
import geopandas as gpd
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import colors

# Load the shapefile
shapefile_path = 'Admin2.shp'  # Replace with the actual shapefile name
india_states = gpd.read_file(shapefile_path)

# Load the CSV file
csv_path = 'Rainfall_Expectation_rev.csv'  # Replace with the actual CSV file name
rainfall_data = pd.read_csv(csv_path)

# Merge shapefile and CSV data
india_states = india_states.merge(rainfall_data, left_on='ST_NM', right_on='ST_NM', how='left')

# Add a column for classification
india_states['Category'] = india_states['Rainfall_Expectation'].apply(
    lambda x: 'Surplus' if x > 5 else ('Deficit' if x < -5 else 'Normal')
)

# Define a custom colormap
category_colors = {'Surplus': 'green', 'Deficit': 'yellow', 'Normal': 'brown'}
india_states['Color'] = india_states['Category'].map(category_colors)

# Plot the map
fig, ax = plt.subplots(1, 1, figsize=(15, 12))
india_states.boundary.plot(ax=ax, linewidth=0.8, color='gray')  # Plot boundaries
india_states.plot(
    column='Color',  # Use the custom color column
    color=india_states['Color'],
    legend=True,
    ax=ax,
    legend_kwds={'title': 'Rainfall Expectation', 'fontsize': 10, 'loc': 'lower left', 'frameon': False},
)

# Add state names
for idx, row in india_states.iterrows():
    if row['geometry'] is not None:
        centroid = row['geometry'].centroid
        ax.text(centroid.x, centroid.y, row['ST_NM'], fontsize=8, ha='center', color='black')

# Customize the plot
ax.set_title('State-wise Rainfall Expectation', fontsize=18, fontweight='bold')
ax.axis('off')  # Turn off axis
ax.set_facecolor("#f0f0f0")  # Set background color

# Show the plot
plt.tight_layout()
plt.show()
```

```python
import geopandas as gpd
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches  # Import for legend patches

# Load the shapefile
shapefile_path = 'Admin2.shp'  # Replace with the actual shapefile name
india_states = gpd.read_file(shapefile_path)

# Load the CSV file
csv_path = 'Rainfall_Expectation_rev.csv'  # Replace with the actual CSV file name
rainfall_data = pd.read_csv(csv_path)

# Merge shapefile and CSV data
india_states = india_states.merge(rainfall_data, left_on='ST_NM', right_on='ST_NM', how='left')

# Add a column for classification
india_states['Category'] = india_states['Rainfall_Expectation'].apply(
    lambda x: 'Surplus' if x > 5 else ('Deficit' if x < -5 else 'Normal')
)

# Define a custom colormap
category_colors = {'Surplus': 'green', 'Deficit': 'yellow', 'Normal': 'orange'}
india_states['Color'] = india_states['Category'].map(category_colors)

# Plot the map
fig, ax = plt.subplots(1, 1, figsize=(15, 12))
india_states.boundary.plot(ax=ax, linewidth=0.8, color='black')  # Plot boundaries
india_states.plot(
    color=india_states['Color'], ax=ax  # Use the custom color column
)

# Add state names
for idx, row in india_states.iterrows():
    if row['geometry'] is not None:
        centroid = row['geometry'].centroid
        ax.text(centroid.x, centroid.y, row['ST_NM'], fontsize=8, ha='center', color='black')

# Customize the plot
ax.set_title('State-wise Rainfall Expectation', fontsize=18, fontweight='bold')
ax.axis('off')  # Turn off axis
ax.set_facecolor("#f0f0f0")  # Set background color

# **Add Legend**
legend_patches = [
    mpatches.Patch(color=category_colors['Surplus'], label='Surplus'),
    mpatches.Patch(color=category_colors['Deficit'], label='Deficit'),
    mpatches.Patch(color=category_colors['Normal'], label='Normal'),
]
ax.legend(handles=legend_patches, title='Rainfall Expectation', fontsize=10, loc='lower left', frameon=False)

# Show the plot
plt.tight_layout()
plt.show()
```
