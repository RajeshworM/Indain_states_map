```python
import geopandas as gpd
import matplotlib.pyplot as plt

# Path to the shapefile
shapefile_path = r"C:\Users\IEG\Desktop\state_map\Admin2.shp"

# Load the shapefile using geopandas
gdf = gpd.read_file(shapefile_path)

# Print the available column names to inspect the data
print("Available columns in the shapefile:\n", gdf.columns)

# Choose the correct column for state names
STATE_COLUMN = 'ST_NM'  # Replace with the correct column name after inspection

# Plot the map with a colormap for different states
fig, ax = plt.subplots(1, 1, figsize=(15, 10))

# Create a colorful map with the legend outside the map
gdf.plot(
    ax=ax, 
    column=STATE_COLUMN, 
    cmap='tab20', 
    legend=True, 
    edgecolor='black',
    legend_kwds={
        'bbox_to_anchor': (1.05, 1),  # Position the legend outside the map
        'loc': 'upper left',          # Place the legend at the upper left of the bbox
        'fontsize': 'small'           # Adjust the legend font size
    }
)

# Add a title to the map
ax.set_title("Indian States Map", fontsize=15)

# Annotate each state with its name
for x, y, label in zip(gdf.geometry.centroid.x, gdf.geometry.centroid.y, gdf[STATE_COLUMN]):
    ax.text(x, y, label, fontsize=8, ha='center', va='center')

# Remove the axis for a cleaner map
ax.set_axis_off()

# Adjust layout to fit the legend outside the map
plt.tight_layout()

# Show the map
plt.show()
```
